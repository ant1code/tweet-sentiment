```
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('BERT', 'dBERT 2', 'dBERT 3', 'R-forests', 'Tom', 'Jessica')
y_pos = np.arange(len(objects))
performance = [0.7903,0.7890,0.7845,0.7005,0.6415,0.6333]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Correctly answered')
plt.title('Comparision')
axes = plt.gca()
axes.set_ylim([0.55,0.85])
rects = axes.patches
for rect, label in zip(rects, performance):
    height = rect.get_height()
    axes.text(rect.get_x() + rect.get_width() / 2, height, label,
            ha='center', va='bottom')
plt.show()
```
